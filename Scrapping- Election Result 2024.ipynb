```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def fetch_eci_results(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.content
    else:
        print(f"Failed to fetch data from {url}")
        return None


def parse_eci_results(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    state_blocks = soup.find_all('div', class_='state-item')
    election_data = []
    
    for state_block in state_blocks:
        state_name = state_block.find('h2').text.strip()
        constituencies = state_block.find('h1').text.strip()
        url = state_block.find('a')['href']
        
        state_info = {
            'State': state_name,
            'Constituencies': constituencies,
            'URL': url
        }
        election_data.append(state_info)
    
    return election_data

def fetch_and_parse_detail(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        result_details = soup.find('div', class_='result-details')
        if result_details:
            detailed_info = result_details.text.strip()
        else:
            detailed_info = "No detailed information found."
        return detailed_info
    else:
        print(f"Failed to fetch data from {url}")
        return None

def main():
    url = 'https://results.eci.gov.in'
    html_content = fetch_eci_results(url)
    
    if html_content:
        election_data = parse_eci_results(html_content)
        
        if election_data:
            for state_info in election_data:
                state_name = state_info['State']
                state_url = state_info['URL']
                print(f"State: {state_name}, URL: {state_url}")
                # Fetch and parse details for each state's URL
                detailed_info = fetch_and_parse_detail(state_url)
                print(f"Detailed information for {state_name}:")
                print(detailed_info)
                print("\n")
                
            df = pd.DataFrame(election_data)
            print("\nDataFrame with election results data:")
            print(df)
            
            df.to_csv('election_results_data.csv', index=False)
            print("\nData saved to 'election_results_data.csv'")
    else:
        print("Failed to fetch HTML content.")

if __name__ == "__main__":
    main()
```
